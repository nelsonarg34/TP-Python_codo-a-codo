```python
num = int(input("Ingrese un número entero: "))
divisores =[]
for divisor in range(1,num+1):
    if num%divisor == 0:
        divisores.append(divisor)
div_str = str(divisores)[1:-1]
print("Es divisible por ",div_str)
```

```python
def esPrimo(num):
    if num<1:
        return False
    elif num==2:
        return True
    else:
        for divisor in range(2,num+1):
            if num%divisor==0:
                return False
            else:
                return True

num=int(input("Ingrese un número: "))
resultado=esPrimo(num)
if resultado is True:
        print("Es primo")
else:
    print("No es primo")
```

```python
num=int(input("Ingrese una nota (1-10): "))
while num>=0 and num<=10:
    if num>=0 and num<=3:
        print("Muy deficiente")
    if num>3 and num<=5:
        print("Insuficiente")
    if num>5 and num<=6:
        print("Suficiente")
    if num>6 and num<=7:
        print("Bien")
    if num>7 and num<=9:
        print("Notable")
    if num>9 and num<=10:
        print("Sobresaliente")
    num=int(input("Ingrese otra nota (Salir = 11): "))   
```
